```python
class CustomCallback(keras.callbacks.Callback):
    def on_train_begin(self, logs=None):
        send_message("Starting training")

    def on_train_end(self, logs=None):
        send_message("Stop training")

    def on_epoch_begin(self, epoch, logs=None):
        send_message("Start epoch {} of training".format(epoch))

    def on_epoch_end(self, epoch, logs=None):
        send_message("End epoch {} of training".format(epoch))
        
    def on_test_begin(self, logs=None):
        send_message("Start testing")

    def on_test_end(self, logs=None):
        send_message("Stop testing")

    def on_predict_begin(self, logs=None):
        send_message("Start predicting")

    def on_predict_end(self, logs=None):
        send_message("Stop predicting")
        
    def on_train_batch_begin(self, batch, logs=None):
        send_message("Training start of batch {}".format(batch))

    def on_train_batch_end(self, batch, logs=None):
        send_message("Training end of batch {}".format(batch))

    def on_test_batch_begin(self, batch, logs=None):
        send_message("Evaluating start of batch {}".format(batch))

    def on_test_batch_end(self, batch, logs=None):
        send_message("Evaluating end of batch {}".format(batch))

    def on_predict_batch_begin(self, batch, logs=None):
        send_message("Predicting start of batch {}".format(batch))

    def on_predict_batch_end(self, batch, logs=None):
        send_message("Predicting end of batch {} ".format(batch))
```

```python
cb = CustomCallback() #Creating the Callback
```

```python
x,y = make_classification(n_samples=1000, n_classes=2, n_features=20)
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.3)
```

```python
def create_model():
    model = Sequential()
    model.add(Dense(60, input_dim=20, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(loss="mean_squared_error", optimizer=keras.optimizers.RMSprop(learning_rate=0.1), metrics=["mean_absolute_error"])

    return model

model = create_model()
```

```python
model.fit(x_train,y_train,batch_size=32, epochs=20, callbacks=[cb], verbose=1) #Fitting the model
```

```python
res = model.evaluate(x_test, y_test, batch_size=32, verbose=1, callbacks=[cb])
```

```python
res = model.predict(x_test, batch_size=32, callbacks=[cb])
```
